# Criação de Matriz Origem-Destino Rodoviária

```python
from selenium import webdriver
from selenium.webdriver.edge.service import Service as EdgeService
from selenium.webdriver.edge.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.keys import Keys
import time
import pandas as pd
```

```python
od_rodo_origens = pd.read_csv('data/csv/od_rodo_origens.csv', delimiter = ',')
```

```python
od_rodo_origens.columns
```

```python
def criar_pares_origem_destino(od_rodo_origens):
    # Lista para armazenar os novos pares de origem-destino e seus códigos
    origens = []
    destinos = []
    cod_origens = []
    cod_destinos = []

    # Conjunto para armazenar origens únicas processadas
    origens_unicas = set()

    # Obter o número total de municípios únicos (origens)
    total_municipios = len(od_rodo_origens['origem'].unique())

    # Criar todos os pares possíveis de origem e destino
    for i, origem_row in od_rodo_origens.iterrows():
        for j, destino_row in od_rodo_origens.iterrows():
            # Verificar se origem e destino não são os mesmos
            if origem_row['cod_origem'] != destino_row['cod_origem']:
                # Adicionar o par origem-destino
                origens.append(origem_row['origem'])
                destinos.append(destino_row['origem'])
                cod_origens.append(origem_row['cod_origem'])
                cod_destinos.append(destino_row['cod_origem'])
                
                # Adicionar a origem ao conjunto de origens únicas
                origens_unicas.add(origem_row['origem'])

        # Mostrar o progresso com base nas origens processadas
        progresso = len(origens_unicas)
        print(f"Progresso: {progresso} de {total_municipios} municípios de origem únicos processados")

    # Criar um novo DataFrame com as colunas preenchidas
    od_rodo = pd.DataFrame({
        'origem': origens,
        'cod_origem': cod_origens,
        'destino': destinos,
        'cod_destino': cod_destinos
    })

    return od_rodo

# Executando a função
od_rodo = criar_pares_origem_destino(od_rodo_origens)
```

```python
od_rodo.to_csv('data/csv/od_rodo.csv', index=False)
```
